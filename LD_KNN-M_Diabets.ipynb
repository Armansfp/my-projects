## Determine the Features & Target Variable

```python
X= df.drop('class', axis=1)
y=df['class']
```

## Split the Dataset to Train & Test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Scaling the Features



```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train=scaler.transform(X_train)
scaled_X_test=scaler.transform(X_test)
```

# Train the Model using the X_train and y_train

## K-NN Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier(n_neighbors=1)
```

```python
knn_model.fit(scaled_X_train ,y_train)
```

```python
#Predicting the Test Data
y_pred=knn_model.predict(scaled_X_test)
```

```python
#We want to khonw the Prediction Value and compare them to the Actual Value.
pd.DataFrame({'Y_test':y_test, 'Y_pred':y_pred}).head()
```

### Evaluating the Model

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn import metrics
```

```python
accuracy_score(y_test , y_pred)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

### Elbow Method for Choosing Reasonable K Values

```python
test_error_rate= []


for k in range (1, 30):
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(scaled_X_train, y_train)
    
    y_pred_test = knn_model.predict(scaled_X_test)
    
    test_error=1- accuracy_score(y_test, y_pred_test)
    test_error_rate.append(test_error)
```

```python
test_error_rate
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, 30), test_error_rate, label='Test Error')
plt.legend()
plt.ylabel('Error Rate')
plt.xlabel('K Value')
```

### Creating a Pipeline to find K value

```python
scaler= StandardScaler()
```

```python
knn= KNeighborsClassifier()
```

```python
knn.get_params().keys()
```

```python
operations= [('scaler', scaler), ('knn', knn)]
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe= Pipeline(operations)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
k_values= list(range(1, 20))
```

```python
param_grid= {'knn__n_neighbors': k_values}
```

```python
full_cv_classifier= GridSearchCV(pipe, param_grid, cv=5, scoring='accuracy')
```

```python
full_cv_classifier.fit(X_train, y_train)
```

```python
full_cv_classifier.best_estimator_.get_params()
```

```python
full_cv_classifier.cv_results_.keys()
```

```python
#Final Model
knn_model= KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
k_range = list(range(1, 30))
param_grid = dict(n_neighbors=k_range)
  
# defining parameter range
grid = GridSearchCV(knn_model, param_grid, cv=10, scoring='accuracy', return_train_score=False,verbose=1)
  
# fitting the model for grid search
grid_search=grid.fit(X_train, y_train)
```

```python
print(grid_search.best_params_)
```

```python
#Checking Accuracy on Test Data

knn_model= KNeighborsClassifier(n_neighbors=21)

knn_model.fit(scaled_X_train, y_train)

y_pred= knn_model.predict(scaled_X_test)
test_accuracy=accuracy_score(y_test,y_pred)*100

print("Accuracy for our testing dataset with tuning is : {:.2f}%".format(test_accuracy) )
```
