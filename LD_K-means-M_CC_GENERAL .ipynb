# Dealing with Outliers

this technique is used when you can consider that all outliers express the same behaviours or patterns , meaning the model wouldn't learn anything new by allowing the outliers to remain.

# **Balance column**

```python
upper_limit = df['BALANCE'].mean() + 3*df['BALANCE'].std()
print(upper_limit)
lower_limit = df['BALANCE'].mean() - 3*df['BALANCE'].std()
print(lower_limit)
```

```python
#Apply the limits to feature.
df['BALANCE'] = np.where(df['BALANCE']>upper_limit , upper_limit, np.where(df['BALANCE']<lower_limit, lower_limit, df['BALANCE']))
```

```python
#Find the outliers by using IQR and replace with the mean value.
def impute_outliers_IQR(df):
  q1 = df.quantile(0.25)
  q3 = df.quantile(0.75)

  IQR = q3 - q1

  upper = df[~(df>(q3 + 1.5*IQR))].max()
  lower = df[~(df<(q1 - 1.5*IQR))].min()

  df = np.where(df>upper, df.mean(), np.where(df<lower, df.mean(), df))

  return df
```

```python
df['BALANCE'] = impute_outliers_IQR(df['BALANCE'])
df.describe()['BALANCE']
```

```python
sns.boxplot(data=df , y='BALANCE')
```
