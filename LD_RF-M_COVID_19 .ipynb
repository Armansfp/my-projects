#REMOVING OUTLIERS 

**Definition:**

**In Statistics, an outlier is a data point that differs significantly from other observations (Grubbs, 1969; Maddala, 1992).**

**An outlier may be due to variability in the measurement or it may indicate experimental error, the latter are sometimes excluded from the data set (Grubbs, 1969).**

**An outlier can couse serious problem in statistical analysis.**

* [Wikipedia Article](http://en.wikipedia.org/wiki/Outlier)


###  **Remove the outliers**

```python
#### LDH ####

upperBoundB = []
for i in range(3):
    q1 = df[df['TARGET']==i]['LDH'].quantile(0.25)
    q3 = df[df['TARGET']==i]['LDH'].quantile(0.75)

    iqr = q3 - q1

    upperBoundB.append(q3 +(1.5 * iqr))

############################################################
fig = plt.figure( figsize=(8, 6))
ax = fig.add_axes([0,0,1,1])
sns.boxplot(ax=ax, data=df, x='TARGET', y='LDH',flierprops=dict(marker='o', markersize=6),fliersize=2)


ax.axhline(y=upperBoundB[0],color='b')
ax.axhline(y=upperBoundB[1],color='r')
ax.axhline(y=upperBoundB[2],color='y')

############################################################

df = df.drop(index= df[ df['TARGET'] == 0 ]['LDH'][ df[ df['TARGET'] == 0 ]['LDH'] > upperBoundB[0]].index
              , axis=0)

df = df.drop(index= df[ df['TARGET'] == 1 ]['LDH'][ df[ df['TARGET'] == 1 ]['LDH'] > upperBoundB[1]].index
              , axis=0)

df = df.drop(index= df[ df['TARGET'] == 2 ]['LDH'][ df[ df['TARGET'] == 2 ]['LDH'] > upperBoundB[2]].index
              , axis=0)
```

## Creating Age Group Feature

-  Infant = 0-1 year.
-  Toddler = 2-4 yrs.
-  Child = 5-12 yrs.
-  Teen = 13-19 yrs.
-  Adult = 20-39 yrs.
-  Middle Age Adult = 40-59 yrs.
-  Senior Adult = 60+

<div class="alert alert-block alert-warning">
We merge the four first groups because of the range of 'AGE' column into one group
</div>


```python
#Creating age group feature.

def ageGroups(x):
  if(x>=0 and x<=19):
    return 'Teen'
  elif(x>=20 and x<=39):
    return 'Adults'
  elif(x>=40 and x<=59):
    return 'Middle Age Adult'
  elif(x>=60):
    return 'Senior Adults'

df['ageGroups'] = df['AGE'].transform(ageGroups)
```

```python
#### AST ###
#### No Different for women and men
#### Different for age groups

print('AST NULL Value : ', df['AST'].isnull().sum())
df['AST'] = df.groupby(by='ageGroups')['AST'].transform(lambda x : x.fillna(x.mean()))
print('AST NULL Value : ', df['AST'].isnull().sum())
```

### ***B. Working base on Rows Missing Data***

**If only a few rows are missing some values, then it might just be a good idea to drop those rows.** 

**What does this cost you in terms of performace? It essentialy removes potential training/testing data, but if its only a few rows, its unlikely to change performance.**


**Sometimes it is a good idea to remove a feature entirely if it has too many null values. However, you should carefully consider why it has so many null values, in certain situations null could just be used as a separate category.** 

(Take for example a feature column for the number of cars that can fit into a garage. Perhaps if there is no garage then there is a null value, instead of a zero. It probably makes more sense to quickly fill the null values in this case with a zero instead of a null. Only you can decide based off your domain expertise and knowledge of the data set!)

**every Feature with missing data must be checked!**

**We choose a threshold of 1%. It means, if there is less than 1% of a feature are missing,then we will consider just dropping that rows**

### ** Dropping Rows:**

```python
# As we can see here, the perecntage of the missing values in 'WBC', 'Piastrine', and 'AST' is less than 1%.
# So, we can drop the rows containg missing values corresponding these columns.
# But in this dataset, it is better to avoid dropping them, as we did 
# In the below snippet code shows how to drop these rows


missValue = missingPercentage(df)

print('-------')

print("The perecntage of the missing values in 'WBC', 'Piastrine', and 'AST' before dropping rows\n ")
print(missValue[( missValue<1) & (missValue>0)])
print('-------')

print("The perecntage of the missing values in 'WBC', 'Piastrine', and 'AST' after dropping rows\n ")
df.dropna(subset=['WBC','Piastrine','AST'],axis=0)[['WBC','Piastrine','AST']].isnull().sum()

```
